**Uploading Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ev_charging_raw_data.csv')
df.head()
```

**Cleaning and Transforming Data**

```python
df.info()
```

```python
df.describe()
```

**Rounding Float Values to 2 Decimal Points**

```python
float_columns = df.select_dtypes(include=['float64']).columns
df[float_columns] = df[float_columns].round(2)
print (df[['Daily_Avg_Utilization_Hrs','Electricity_Cost_INR','Install_Cost_INR']].head())
```

**Correcting the Spelling Incorrections**

```python
Charger_replace = {'l2' : 'L2', 'dcfc' : 'DCFC'}
df['Charger_Type'] = df['Charger_Type'].str.strip().replace(Charger_replace)
print (df['Charger_Type'].value_counts())

City_replace = {'BANGALORE' : 'Bangalore', 'hydrabad' : 'Hyderabad', 'Coimbator' : 'Coimbatore', 'chennai':'Chennai'}
df['City'] = df['City'].str.strip().replace(City_replace)
print (df['City'].value_counts())
```

**Dropping Rows with Zeros and Converting Negatives**

```python
df['Install_Cost_INR'] = df['Install_Cost_INR'].abs()
df['Electricity_Cost_INR'] = df['Electricity_Cost_INR'].abs()
df['Avg_Charging_Rate_INR'] = df['Avg_Charging_Rate_INR'].abs()
df['Total_Energy_Dispensed_kWh'] = df['Total_Energy_Dispensed_kWh'].abs()
df['Total_Maintenance_Cost_INR'] = df['Total_Maintenance_Cost_INR'].abs()
df['Daily_Avg_Utilization_Hrs'] = df['Daily_Avg_Utilization_Hrs'].abs()
df = df[df['Total_Energy_Dispensed_kWh']!=0]
df = df[df['Total_Maintenance_Cost_INR']!=0]
df = df[df['Daily_Avg_Utilization_Hrs']!=0]
df = df[df['Avg_Charging_Rate_INR']!=0]
df = df[df['Install_Cost_INR']!=0]
df = df[df['Electricity_Cost_INR']!=0]
df.describe()
```

**Handling Null Values**

```python
print (df.isnull().sum())
```

**Replacing NULL**

```python
#Removing Error Rows and Replacing NULL with Mean
df = df[df['Daily_Avg_Utilization_Hrs']<=24]
df['Daily_Avg_Utilization_Hrs'] = df['Daily_Avg_Utilization_Hrs'].fillna(df['Daily_Avg_Utilization_Hrs'].mean())

# Replacing NULL with "Unknown"
df['Charger_Type'] = df['Charger_Type'].fillna('Unknown')


#Replacing NULL with Grouped Average
df['Total_Energy_Dispensed_kWh'] = df['Total_Energy_Dispensed_kWh'].fillna(df.groupby(['City','Charger_Type'])['Total_Energy_Dispensed_kWh'].transform('mean')).round(2)


df['Total_Maintenance_Cost_INR'] = df['Total_Maintenance_Cost_INR'].fillna(df.groupby(['City','Charger_Type'])['Total_Maintenance_Cost_INR'].transform('mean')).round(2)
print (df.isnull().sum())
```

**Converting Object to Date format**

```python
df['Install_Date'] = pd.to_datetime(df['Install_Date'], dayfirst=True, format = 'mixed')
df.head()
```

**Creating New Columns**

```python
# Total_Revenue
df['Total_Revenue'] = (df['Avg_Charging_Rate_INR']*df['Total_Energy_Dispensed_kWh']).round(2)

# Total_Operational_Cost
df['Total_Operational_Cost'] = ((df['Total_Energy_Dispensed_kWh']*df['Electricity_Cost_INR']) + df['Total_Maintenance_Cost_INR']).round(2)

# Net_Profit
df['Net_Profit'] = (df['Total_Revenue'] - df['Total_Operational_Cost']).round(2)

# ROI
df['ROI'] = ((df['Net_Profit']/df['Install_Cost_INR'])*100).round(2)
```

**Exporting the Cleaned DataSet**

```python
df.to_csv("ev_charging_cleaned_data.csv", index = False)
```
